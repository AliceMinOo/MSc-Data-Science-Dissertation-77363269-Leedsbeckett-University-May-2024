```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Define tasks with start and end dates
sprint_1_tasks = [
    {'task': 'Introduction', 'start_date': '2024-01-01', 'end_date': '2024-01-03'},
    {'task': 'Aim and Objective', 'start_date': '2024-01-03', 'end_date': '2024-01-07'},
    {'task': 'Rationale', 'start_date': '2024-01-07', 'end_date': '2024-01-11'},
    {'task': 'Literature Review', 'start_date': '2024-01-11', 'end_date': '2024-01-17'},
    {'task': 'Methodology', 'start_date': '2024-01-17', 'end_date': '2024-01-22'},
    {'task': 'Ethical Consideration', 'start_date': '2024-01-22', 'end_date': '2024-01-23'},
    {'task': 'Planning', 'start_date': '2024-01-23', 'end_date': '2024-01-24'}

]

sprint_2_tasks = [
    {'task': 'Data collection', 'start_date': '2024-02-01', 'end_date': '2024-02-03'},
    {'task': 'Data Preprocessing', 'start_date': '2024-02-03', 'end_date': '2024-02-13'},
    {'task': 'Model Implementation', 'start_date': '2024-02-13', 'end_date': '2024-03-13'}
]

sprint_3_tasks = [
    {'task': 'Model Evaluation', 'start_date': '2024-03-13', 'end_date': '2024-04-07'},
    {'task': 'Comparative Analysis of Models', 'start_date': '2024-04-07', 'end_date': '2024-04-23'}
]

sprint_4_tasks = [
    {'task': 'Abstract', 'start_date': '2024-04-23', 'end_date': '2024-04-24'},
    {'task': 'Result', 'start_date': '2024-04-24', 'end_date': '2024-04-28'},
    {'task': 'Discussion', 'start_date': '2024-04-28', 'end_date': '2024-04-30'},
    {'task': 'Conclusion', 'start_date': '2024-04-30', 'end_date': '2024-05-02'}
]
# Combine tasks of sprints
tasks = sprint_1_tasks + sprint_2_tasks  + sprint_3_tasks + sprint_4_tasks

# Convert dates to datetime objects
for task in tasks:
    task['start_date'] = mdates.datestr2num(task['start_date'])
    task['end_date'] = mdates.datestr2num(task['end_date'])

# Set figure size
plt.figure(figsize=(10, 6))  # Width: 10 inches, Height: 6 inches

# Create Gantt chart
fig, ax = plt.subplots(figsize=(10, 6))

for i, task in enumerate(tasks):
    ax.barh(y=i, width=task['end_date'] - task['start_date'], left=task['start_date'], height=0.5, align='center', label=task['task'])

# Set y-ticks as task names
ax.set_yticks(range(len(tasks)))
ax.set_yticklabels([task['task'] for task in tasks])

# Format x-axis as dates
ax.xaxis_date()

# Add labels and title
plt.xlabel('Date')
plt.ylabel('Tasks')
plt.title('Agile Project Timeline')

# Set date format on x-axis
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))

# Display the plot
plt.show()
```
